# Creating a List

# Accessing Items from a List

```python
# Indexing
L = [1,2,3,4,5]
print(L[0])
L1 = [1,2,3,[4,5]]
print(L1[3][0])
print(L1[3][1])
# Slicing
L2 = [1,2,3,4,5,6]
print(L2[0:])
print(L2[4:6])
print(L2[:])
print(L2[::-1])
print(L2[-1:-7:-1])
print(L2[0::2])
print(L2[-5:-3:2])
```

```python
L = [1,2,3,4,5]
L[4] = 500
print(L)
L[1:4] = [200,300,400]
print(L)
print(type(L[0]))
print(type(L[2]))
print(type(L[4]))
```

# Operations on Lists
- Arithmetic
- Membership
- Loop

```python
# Arithmetic (+, *)
L1 = [1,2,3,4]
L2 = [5,6,7,8]
print(L1 + L2)
print(L1*3)
```

```python
# Membership
L1 = [1,2,3,4,5]
L2 = [1,2,3,4,[5,6]]
print(5 in L1)
print(6 in L2)
print(5 in L2)
```

```python
# Loops
L1 = [1,2,3,4,5]
L2 = [1,2,3,4,[5,6]]

for i in L1:
  print(i)
```

```python
# len, min, max, sorted
L = [2,1,5,7,0]

print(len(L))
print(min(L))
print(max(L))
print(sorted(L))
print(sorted(L, reverse = True))
```

```python
L = [1,2,1,3,4,1,5]
print(L.count(1))
print(L.count(2))
```

```python
# reverse
L1 = [1,2,3,4,5,6]
L1.reverse()
print(L1)
```

```python
# sort vs (sorted)
L = [2,1,5,7,0]
print(L)
print(sorted(L))
print(L)
L.sort()
print(L)
# .sort() is permanent
```

```python
# copy
L = [2,1,5,7,0]
print(id(L))
L1 = L.copy()
print(id(L1))
```

# List Comprehension

List Comprehension provides a concise way of creating this.

newlist = [expression for item in iterable if condition == True]

Advantages of list comprehension: -
- More time-efficient and space-efficient than loops.
- Require fewer lines of code
- Transforms iterative statement into a formula

```python
# Add 1 to 10 numbers in a list.
L = [num for num in range(1,11)]
print(L)
```

```python
# Scalar multiplication on vector
v = [2,3,4]
s = -3

L = [s*i for i in v]
print(L)
```

```python
# Add Squares
L = [1,2,3,4,5]
square = [num**2 for num in L]
print(square)
```

```python
# Print all the numbers divisible by 5 in the range of 1 to 50
L = [num for num in range(1,51) if num%5==0]
print(L)
```

```python
# find languages which start with letter p
languages = ['java','python','php','c','javascript']

l = [language for language in languages if language.startswith('p')]
print(l)
```

```python
# Nested if with list comprehension
basket = ['apple', 'guava','cherry','banana']
my_fruits = ['apple','kiwi','grapes','banana']

# add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'.

li = [fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
print(li)
```

```python
# Print a (3X3) matrix using list comprehension -> Nested list comprehension
matrix = [[i*j for i in range(1,4)] for j in range(1,4)]
print(matrix)
```

```python
# cartesian products -> List comprehension on 2 lists together
L1 = [1,2,3,4]
L2 = [5,6,7,8]
cartesian_product = [(x*y) for x in L1 for y in L2]
print(cartesian_product)
```

# 2 Ways to traverse a list
- itemwise
- indexwise

```python
# itemwise
L = [1,2,3,4]

for i in L:
  print(i)
```

```python
# indexwise
L = [1,2,3,4]

for i in range(0,len(L)):
  print(i)
```

# Zip Function
The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together and then the second item in each passed iterator are paired together.
If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator.

```python
L1 = [1,2,3,4]
l2 = [-1,-2,-3,-4]
list(zip(L1,l2))
[i+j for i, j in zip(L1,l2)]
```
